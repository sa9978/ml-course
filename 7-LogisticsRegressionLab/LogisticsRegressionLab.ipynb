```python
X = np.array(data[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip']])
X[0:5]
```

```python
Y = np.array(data['churn'])
Y[0:5]
```

In the machine learning algorithms if the values of the features are closer to each other there are chances for the algorithm to get trained well and faster.
So we scale the data, to make sure they are close enough.

```python
from sklearn import preprocessing

scalar = preprocessing.StandardScaler().fit(X)
X = scalar.transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print("x test shape:", np.shape(x_test))
print("x train shape:", np.shape(x_train))
print("y test shape:", np.shape(y_test))
print("y train shape:", np.shape(y_train))
```

## Logistics Regression Parameters:

- ### C:
It represents the inverse of regularization strength. It should be a positive float. Usually search logarithmically: [.001, .01, .1, 1, 10, 100, 1000].
For **small values of C**, we increase the regularization strength which will create simple models which **underfit** the data. For **large values of C**, we low the power of regularization which imples the model is allowed to increase it's complexity, and therefore, **overfit** the data.
- ### solver:
Algorithm to use in the optimization problem.
  - #### newton-cg
   A newton method. Newton methods use an exact Hessian matrix. It's **slow for large datasets**, because it computes the second derivatives.
  - #### lbfgs
  Stands for Limited-memory Broyden–Fletcher–Goldfarb–Shanno. It approximates the second derivative matrix updates with gradient evaluations. It stores only the last few updates, so it saves memory. It **isn't super fast with large data sets**. It will be the default solver as of Scikit-learn version 0.22.0.
  - #### liblinear
  It's a good choice for **small datasets**.
  - #### sag
  Stochastic Average Gradient descent. A variation of gradient descent and incremental aggregated gradient approaches that uses a random sample of previous gradient values. **Fast for big datasets**.
  - #### saga
  Extension of sag that also allows for L1 regularization. Should generally train **faster than sag**.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 10, 100, 1000],
              'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
lr = LogisticRegression()
clf = GridSearchCV(lr, param_grid)
clf.fit(x_train, y_train)
print("best parameters", clf.best_params_)
print("best score", clf.best_score_)
```

```python
y_test_ = clf.predict(x_test)
print(y_test[0:5])
print(y_test_[0:5])
```

**predict_proba** gives the probability of each y in each class. It Chooses the class with larger value.

```python
y_test_prob = clf.predict_proba(x_test)
print(y_test_prob[0:5])
```

## Evaluation

```python
from sklearn.metrics import jaccard_score
print("jaccard_scor 1:", jaccard_score(y_true=y_test, y_pred=y_test_, pos_label=1))
print("jaccard_scor 0:", jaccard_score(y_true=y_test, y_pred=y_test_, pos_label=0))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf, x_test, y_test)
```
